```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

Plotting the vector.

```python
# quiver is the case which holds the arrow.
plt.quiver(0,0,4,5)
plt.show()
```

In the above graph, the cor-ordinates are not mentiond properly, so we have to solve this.

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1, color='b')
# This parameter controls how the length of the arrow is scaled, based on the coordinate system.
# This parameter controls how the direction (angle) of the arrow is interpreted.
# Giving the lower and higher limit for x and y axis.
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,-5,-6, scale_units='xy', angles='xy', scale=1, color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Addition of two vectors

```python
# Vectors are arrays, so we are creating a numpy array
vector1= np.asarray([0,0,2,3])
vector2= np.asarray([0,0,-3,2])
sum = vector1+vector2
print(sum)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,-3,2, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-1,5, scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Subtraction of Two Vectors

```python
vector1= np.asarray([0,0,2,3])
vector2= np.asarray([0,0,-3,2])
sub = vector1-vector2
print(sub)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,-3,2, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,5,1, scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Multiplying a vector by scalar

```python
vector1 = np.asarray([0,0,2,3])
vector2 = 2 * vector1
print(vector2)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0,0,4,6, scale_units='xy', angles='xy', scale=1, color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Dot product of 2 vectors

```python
# Instead of 2 values, we can also take 3 values.
a = np.array([2,3])
b = np.array([4,4])
a_dot_b = np.dot(a,b)
```

```python
print(a_dot_b)
```

Cross Product of 2 vectors

```python
a = np.array([2,3])
b = np.array([4,4])
a_cross_b = np.cross(a,b)
```

```python
print(a_cross_b)
```

Projection of 'a'  vector on 'v' vector.

```python
a = np.asarray([2,4])
v = np.asarray([4,-5])
# Finding Magnitude of 'v' vector:
# here i have used sum previously, so i was egtting error.
# so i have written np.sum, which resolves it
mag_v= np.sqrt(np.sum(v**2))
# here we are first taking square of both values of v
# Then we are adding those both values
# Then we take square root.

proj_of_a_on_v = (np.dot(a,v)/(mag_v)**2)*v
print(proj_of_a_on_v)
```
